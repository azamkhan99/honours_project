```python
import geojson
import pandas as pd
import utm
from shapely import geometry
from shapely import ops
import numpy as np
```

```python
road_list = ['residential','living_street','tertiary','trunk','secondary','primary','pedestrian','tertiary_link','trunk_link','primary_link','secondary_link']

road_dict = {k: v for v, k in enumerate(road_list)}
```

```python
road_dict
```

```python
def load_map(path):
    
        
    with open(path, encoding="utf-8") as f:
        lines = geojson.load(f)
    Roads = []
    for allFeatures in lines.features:
        if 'highway' in allFeatures['properties']:
            roadinfo = allFeatures['properties']
            
            roadinfo['location'] = allFeatures['geometry']['coordinates']

            Roads.append(roadinfo)
    Roads = pd.DataFrame.from_dict(Roads)
    Roads = Roads[Roads.highway.isin(['residential','living_street','tertiary','trunk','secondary','primary','pedestrian','tertiary_link','trunk_link','primary_link','secondary_link'])]
    #Roads = Roads.set_index('id')
    return Roads
```

```python
def utm_convert(row):
    ls = []
    for y in row['location']:
        ls.append(list(utm.from_latlon(y[1],y[0])[0:2]))
    return ls
        
    
```

```python
def linestring(row):
    return geometry.LineString(row.utmLocation)
```

## Leon

```python
path = 'leon.geojson'
with open(path, encoding="utf-8") as f:
    osmlines = geojson.load(f)
osmRoads = []
for allFeatures in osmlines.features:
    if 'highway' in allFeatures['properties']:
        roadinfo = allFeatures['properties']
        
        roadinfo['location'] = allFeatures['geometry']['coordinates']
        
        osmRoads.append(roadinfo)
osmRoads = pd.DataFrame.from_dict(osmRoads)
osmRoads = osmRoads[osmRoads.highway.isin(['residential','living_street','tertiary','trunk','secondary','primary','pedestrian','tertiary_link','trunk_link','primary_link','secondary_link'])]
osmRoads = osmRoads.set_index('id')
```

```python
osmRoads = osmRoads.drop('way/446923114')
osmRoads = osmRoads.drop('way/783220643')
```

```python
osmRoads['utmLocation'] = osmRoads.apply(lambda row: utm_convert(row), axis=1)
```

```python
osmRoads['locationLineString'] = osmRoads.apply(lambda row: linestring(row), axis=1)
```

```python
osmRoads = osmRoads.filter(['name','highway', 'location', 'utmLocation', 'locationLineString'])
```

```python
osmRoads
```

```python
leon = pd.read_csv("leon_walk.csv")
leon = leon.drop('Unnamed: 0', axis=1)
```

```python
def get_roadtype(row, roadmap):
    airspeckCoords=[(row['gpsLatitude'], row['gpsLongitude'])]
    
    closestRoadstoCoords = ''
    airspeckCoordsUtm = [list(utm.from_latlon(*x))[0:2] for x in airspeckCoords]
    airspeckShapelyPoints = [geometry.Point(x) for x in airspeckCoordsUtm]

    for points in airspeckShapelyPoints:
        roadDistancesToPoint = [points.distance(x) for x in roadmap.locationLineString]
        minIdx = np.argmin(roadDistancesToPoint)
        minRoadType = roadmap.highway[minIdx]
        closestRoadstoCoords = minRoadType
        minRoadName = roadmap.name[minIdx]

    return closestRoadstoCoords
```

```python
%%time
leon['roadType'] = leon.parallel_apply(lambda row: get_roadtype(row, osmRoads), axis=1)
```

```python
leon.to_csv("leon_walk.csv")
```
