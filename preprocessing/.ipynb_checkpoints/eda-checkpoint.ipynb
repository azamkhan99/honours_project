```python
import pandas as pd
rd = pd.read_csv('road_dict.csv', header=0).to_dict('records')
road_dict = rd[0]
```

```python
def mean_absolute_percentage_error(y_true, y_pred):
    #y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

# Airspeck-P

```python
peeps = pd.read_csv("peeps_new.csv")
peeps = peeps.drop(['Unnamed: 0', 'closest_pm_id'], axis =1)
peeps = peeps.replace({"roadType": road_dict})
```

```python
mexico = pd.read_csv("mexico.csv")
mexico = mexico.drop(['walk', 'closest_pm_id'], axis =1)
mexico = mexico.replace({"roadType": road_dict})
```

```python
leon = mexico[mexico.city.eq(0)]
guadalajara = mexico[mexico.city.eq(1)]
```

```python
plt.style.use('seaborn')
```

```python
stds = {}
```

```python
stds['Leon\n Personal'] = 14.248043
stds['Guadalajara\n Personal'] = 12.731682
stds['Delhi\n Personal'] = 83.016619
stds['Leon\n Static'] = leon.pm2_5.std()
stds['Guadalajara\n Static'] = guadalajara.pm2_5.std()
stds['Delhi\n Static'] = peeps.pm2_5.std()
```

```python
fig = plt.figure(figsize =(12, 6))
plt.bar(range(len(stds)), list(stds.values()), align='center', color=['tab:blue', 'tab:orange', 'tab:green', 'tab:blue', 'tab:orange', 'tab:green'])
plt.xticks(range(len(stds)), list(stds.keys()), fontsize=12)
plt.ylabel("$PM_{2.5} (µg/m^3)$")
plt.show()
fig.savefig("dataset_stds.png")

```
