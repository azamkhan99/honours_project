# Functions

```python
def closest_prev_timestep_value(row, dataframe):
    uuid1 = row['closest_pm_id']#.item()

    ts = row['timestamp']
    date = str(ts.date())
    
    first = ts - timedelta(minutes=2.5)
    last = ts + timedelta(minutes=2.5)
    first = str(first.time())
    last = str(last.time())

    try:
        preds = dataframe[dataframe.UUID == uuid1].loc[date].between_time(first, last)['pm2_5'].values[0]
        return preds
    except:
        return np.nan
```

```python
def make_datasets(dataframe, static_dataset,city):
    datasets = []
    
    
    for i in range(11,5,-1):
        test = dataframe.copy()
        test['timestamp'] -= timedelta(minutes=i*minutes)
        test = test.drop("closest_pm", axis=1)
        test['closest_pm'] = test.parallel_apply(lambda row: closest_prev_timestep_value(row, static_dataset), axis=1)
        test = test.fillna(method='ffill')
        datasets.append(test)
    
    interleaved = pd.DataFrame(interleave([datasets[0].values, datasets[1].values, datasets[2].values, datasets[3].values, datasets[4].values, datasets[5].values]), columns=['timestamp','pm2_5', 'hour_of_day', 'day_of_week','humidity', 'gpsLongitude', 'gpsLatitude', 'dist_to_closest_pm', 'closest_pm_id', 'roadType', 'closest_pm'])
    interleaved = interleaved[['timestamp','pm2_5', 'hour_of_day', 'day_of_week', 'humidity', 'gpsLongitude', 'gpsLatitude', 'closest_pm', 'dist_to_closest_pm', 'closest_pm_id', 'roadType']]
    
    return interleaved
```

```python
def mean_absolute_percentage_error(y_true, y_pred):
    #y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```
